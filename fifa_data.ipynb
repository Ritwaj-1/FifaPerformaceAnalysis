```python
fifa_df["Weight"]=[int(x.strip("lbs"))  if type(x)==str else x  for x in fifa_df["Weight"] ]    
```

```python
light= fifa_df.loc[fifa_df["Weight"]<125].count()[0]

light_medium=fifa_df.loc[(fifa_df["Weight"]>125) & (fifa_df["Weight"]<150)].count()[0]

medium=fifa_df.loc[(fifa_df["Weight"]>150) & (fifa_df["Weight"]<175)].count()[0]

medium_heavy=fifa_df.loc[(fifa_df["Weight"]>175) & (fifa_df["Weight"]<200)].count()[0]

heavy=fifa_df.loc[fifa_df["Weight"]>=200].count()[0]
```

```python
plt.figure(figsize=(8,6))
weights=[light,light_medium,medium,medium_heavy,heavy]
labels=["Under 125","125-150","150-175","175-200","Over 200"]
explode=[.6,.2,0,0,.6]
plt.title("Weight Distribution of FIFA Players (in lbs)",fontweight="bold",fontsize=12)

plt.pie(weights, labels=labels, autopct="%2f %%",pctdistance=0.8,explode=explode)
plt.show()
```

```python
X=fifa_df["Age"]
Y=fifa_df["Potential"]
```

```python
plt.figure(figsize=(10,6))
sns.set(style="whitegrid")
plt.scatter(X,Y, alpha=0.5)
plt.xlabel("Age",color="Black")
plt.ylabel("Potential",color="Black")
plt.title("Distribution of Potential of Players as their Age in FIFA 2018",color="Black",fontsize=10)
plt.show()
```

```python
X=X.values.reshape(len(X),  1)

# Expected 2D array, got 1D array instead:
# we use this to reshape the array.
# Also pandas series has no attribute reshape , so we use X=X.values.reshape(len(X),  1)
```

```python
X.shape

# checking the shape of the array to be 2-D
```

```python
Y=Y.values.reshape(len(Y),  1)
```

```python
Y.shape
```

```python
regression=LinearRegression()
regression.fit(X,Y)
```

```python
plt.figure(figsize=(10,6))
sns.set(style="whitegrid")
plt.scatter(X,Y, alpha=0.5)
plt.xlabel("Age",color="Black")
plt.ylabel("Potential",color="Black")
plt.title("Distribution of Potential of Players as their Age in FIFA 2018",color="Black",fontsize=10)
plt.plot(X,regression.predict(X),color="Red",linewidth=4)
plt.show()
```

```python
# theta_1

regression.coef_
```

```python
print(regression.score(X,Y))
print((regression.score(X,Y))*100,"%")
```
